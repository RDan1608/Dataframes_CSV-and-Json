## DataFrame Dimension Estudiantes

```python
import pandas as pd
df_Estudiantes= pd.read_csv("Dim_Estudiantes.csv")
df_Estudiantes
```

## DataFrame Dimension Clases

## DataFrame Dimension Carrera

```python
df_Carrera= pd.read_csv("Dim_Carrera.csv")
df_Carrera
```

# Lectura DataFrame JSON

## DataFrame Hechos

```python
df_Hechos= pd.read_json("Hechos.json")
df_Hechos
```

# Union de los Dataframes en un solo DataFrame

```python
Hechos_df = df_Hechos.merge(df_Clases, on='id_Clase').merge(df_Estudiantes, on='Id_Estudiante').merge(
            df_Carrera, on='Id_Carrera').merge(df_Periodo, on='Id_Periodo')
Hechos_df
```

# Manipulaciones de datos 

# ¿Cual es el Rendimiento Academico de Cada Modalidad de Clase?

## Por medio de las notas obtenidas en las diferentes modalidades se sacara un promedio para analizar el rendimiento de los estudiantes en las distintas modalidades de clase.

```python
Hechos_df.groupby(['Modalidad'])[['Notas']].agg(Rendimiento_Academico=('Notas','mean')).round(2)
```

# ¿Cuales son los años con mayor demanda de matricula de estudiantes?

## Consiste en encontrar el numero de estudiantes matriculados por año(Incluyendo los 3 semestres)

```python

Hechos_df['Fecha'] = pd.to_datetime(Hechos_df['Fecha'], format='%Y-%m-%d')
Hechos_df['año'] = Hechos_df['Fecha'].dt.year

df1=Hechos_df.groupby('año',as_index=False)[['Id_Estudiante']].nunique()
df4=df1.groupby('año',as_index=False)[['Id_Estudiante']].agg(Numero_Estudiantes=('Id_Estudiante','max'))
df4
```

# ¿Cual es la carrera con mejores promedios a nivel periodo?

```python
df1=Hechos_df.groupby(['Nombre_Carrera','Id_Periodo'],as_index=False)[['Notas']].agg(Promedio=('Notas','mean')).round(2)
df2=df1.groupby(['Nombre_Carrera'])[['Promedio']].mean()
df2['Promedio'].idxmax()
```

# ¿Cual es el mayor rendimiento academico  que a obtenido cada carrera en base a sus notas aprobadas y años anteriores?

## Se analizara historicamente el rendimiento academico de cada carrera en base a sus clases aprobadas (Nota>65) y años anteriores

```python
import datetime as dt
Hechos_df['Fecha'] = pd.to_datetime(Hechos_df['Fecha'], format='%Y-%m-%d')
Hechos_df['año'] = Hechos_df['Fecha'].dt.year
df1=Hechos_df[Hechos_df['Notas'] > 65]
df2=df1.groupby(['año','Nombre_Carrera'])[['Notas']].agg(Promedio=('Notas','mean')).round(2)
df2.reset_index(inplace=True)
df2.groupby(['Nombre_Carrera'])[['Promedio']].mean().round(2)
```
